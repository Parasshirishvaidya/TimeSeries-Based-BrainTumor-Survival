```python
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
img = nib.load(r"C:\SSI Notes\Paper\Dataset1\CFB-GBM\001\t0\1_t0_adc.nii.gz").get_fdata()
mask = nib.load(r"C:\SSI Notes\Paper\Dataset1\CFB-GBM\001\t0\1_t0_ct.nii.gz").get_fdata()
slice_idx = img.shape[2] // 2
plt.imshow(img[:, :, slice_idx], cmap="gray")
plt.imshow(mask[:, :, slice_idx], alpha=0.4)
plt.title("FLAIR + Tumor Mask")
plt.axis("off")
plt.show()
```
