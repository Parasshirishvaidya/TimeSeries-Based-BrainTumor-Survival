```python
import os
import shutil

base = r"C:\SSI Notes\Paper\Dataset1\CFB-GBM"
img_out = r"C:\SSI Notes\Paper\Dataset1\Dataset\imagesTr"
lab_out = r"C:\SSI Notes\Paper\Dataset1\Dataset\labelsTr"

os.makedirs(img_out, exist_ok=True)
os.makedirs(lab_out, exist_ok=True)

copied = 0
skipped = 0

for pid in os.listdir(base):
    t0 = os.path.join(base, pid, "t0")
    if not os.path.isdir(t0):
        continue

    flair = None
    gtv = None

    for f in os.listdir(t0):
        if f.endswith("_t0_flair.nii.gz"):
            flair = os.path.join(t0, f)
        elif f.endswith("_t0_gtv.nii.gz"):
            gtv = os.path.join(t0, f)

    if flair and gtv:
        shutil.copy2(flair, os.path.join(img_out, f"gbm_{pid}_flair.nii.gz"))
        shutil.copy2(gtv,   os.path.join(lab_out, f"gbm_{pid}_gtv.nii.gz"))
        copied += 1
    else:
        skipped += 1
        print(f"Skipped {pid} (missing flair or gtv)")

print(f"Done. Copied {copied} cases. Skipped {skipped}.")
```
